# Flari: una herramienta para NER y otras tareas de PLN

https://pypi.org/project/flair/

```python
!pip install -q flair
```

```python
from flair.data import Sentence
from flair.nn import Classifier

tagger = Classifier.load('ner')

sentence = Sentence('Lucía Bose nació en Roma.')
tagger.predict(sentence)

print(sentence)
```

También permite realizar análisis de sentimiento:

```python
analyzer = Classifier.load('sentiment')
analyzer.predict(sentence)
print(sentence)
```

```python
sentence = Sentence('Todas las rupturas son dolorosas.')
analyzer.predict(sentence)
print(sentence)
```

```python
sentence = Sentence('Una veces se gana y otras ... se aprende')
analyzer.predict(sentence)
print(sentence)
```

También es posible cargar un modelo transformador para recuperar el embedding de cada palabra: 

En https://flairnlp.github.io/docs/intro, es posible encontrar más información sobre esta herramienta. Por ejemplo, es posible entrenar tu propio modelo NER o ajustarlo a otro dataset. 

Puedes cargar embeddings, por ejemplo, de Glove o de un modelo transformador:

```python
from flair.embeddings import WordEmbeddings
sentence = Sentence('The sky is blue.')

glove_embedding = WordEmbeddings('glove')
glove_embedding.embed(sentence)
for token in sentence:
    print(token)
    print(token.embedding)
```

```python
from flair.embeddings import TransformerWordEmbeddings

embedding = TransformerWordEmbeddings('bert-base-uncased')
sentence = Sentence('The sky is blue.')

embedding.embed(sentence)
for token in sentence:
    print(token)
    print(token.embedding)
```

```python
from flair.embeddings import FlairEmbeddings
flair_embedding_forward = FlairEmbeddings('news-forward')
sentence = Sentence('The grass is green .')
flair_embedding_forward.embed(sentence)
for token in sentence:
    print(token)
    print(token.embedding)
```
