# Stanza 
Es una herramienta para Named Entity Recognition desarrollada por la Universidad de Standford.
https://stanfordnlp.github.io/stanza/ner.html


```python
!pip install -q stanza
```

```python
import stanza
nlp = stanza.Pipeline(lang='es', processors='tokenize,ner')
```

```python
doc = nlp("Isabel Segura Bedmar trabaja en la Universidad Carlos III de Madrid. Ella vive en Leganés.")
print(*[f'entity: {ent.text}\ttype: {ent.type}' for ent in doc.ents], sep='\n')
```

```python
doc = nlp("Una ONG británica acepta 10 millones de un fondo en Jersey vinculado al rey emérito.")
print(*[f'entity: {ent.text}\ttype: {ent.type}' for ent in doc.ents], sep='\n')
```

```python
doc = nlp('Liudmila Vakulenko aún es regidora de Kozacha Lopan, en la frontera con Rusia, tras ser absuelta de colaboracionismo.')
print(*[f'entity: {ent.text}\ttype: {ent.type}' for ent in doc.ents], sep='\n')
```

Podemos trabajar con otros modelos entrenados en otros dominios como el biomédico:

```python
pipe = stanza.Pipeline("en", processors="tokenize,ner", package={"ner": ["ncbi_disease", "ontonotes"]})
```

```python
doc = pipe("My father has diabetes")
print(doc.ents)
```

```python
# doc = pipe("Lysosomal storage disorders (LSDs)—a group of rare genetic conditions caused by enzyme deficiencies")
doc = pipe('Triple-negative breast cancer (TNBC) accounts for about 10-15% of all breast cancers.')
print(doc.ents)
```
