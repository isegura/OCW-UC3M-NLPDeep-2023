# GPT-2
GPT-2 es un modelo trasnformer auto-regresivo, que utiliza un modelado de lenguaje causal (CLM), es decir, el objetivo es predecir la siguiente palabra en una oración.

Como BERT, GPT-2 también representa la posición absoluta de cada token, por tanto, es aconsejable que el padding se utliza a la derecha. GPT-2 fue pre-entrenado sin padding. 


En particular, las entradas para el modelo son secuencias de texto continuo, mientras que los objetivos son la misma secuencia, desplazada un token a la derecha. Internamente, el modelo utiliza un mecanismo de máscara. Para predecir el token i, el modelo solo usa los tokens anteriores, pero no los tokens futuros.

Vamos a usar el modelo directamente para predecir el siguiente token a una secuencia:

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')
set_seed(42) 
generator("The researcher has presented a", max_length=30, num_return_sequences=5)
```

En lugar de usar el pipeline, podemos usar las clases GPT2Tokenizer y GPT2Model:

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# load tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# we encode the text
text = "The researcher has presented a"
encoded_input = tokenizer(text, return_tensors='pt')

# we apply the model on the output
output = model.generate(input_ids=encoded_input.input_ids)
# print(output[0])

result = tokenizer.decode(output[0])
print(result)
```
