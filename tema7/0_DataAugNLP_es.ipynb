## ¿Qué es Data Augmentation?

**Data augmentation**  son técnicas dedicadas para **aumentar la cantidad de datos**, mediante el uso de diferentes modificaciones que nos permitiran incrementar la cantidad cantidad de ejemplos en el conjunto de datos original. Data augmentatio, además de aumentar el conjunto de datos, también incrementa la diversidad del conjunto de datos. Al entrenar modelos de aprendizaje automático, un mayor número de ejemplos como un regularizador y ayuda a evitar el sobreajuste.

Data augmentation resultan útiles en dominios como la visión artificial y la PLN. Por ejemplo, en visión artificial, las modificaciones que se realizan son : recorte, ampliaciones, giros y rotaciones, cambios de tamaño y de color, etc.

## Data Augmentation for PLN
Source: https://neptune.ai/blog/data-augmentation-nlp

Como bien se sabe, el aprendizaje automático requiere datos anotados correctos para que sus modelos sean capaces de proporcionar buenos resultados. Pero adquirir y etiquetar estos datos es un proceso muy costoso y lento.

**Data augmentation** nos permiten **generar** **datos sintéticos** adicionales a partir de los datos disponibles. 

En visión artificial, estas técnicas (recorte,
voltear, hacer zoom, rotar, etc.) se pueden implementar fácilmente mediante la transformación que se realiza sobre la marcha utilizando generadores de datos, sin necesidad de preparar nada más antes del entrenamiento. Es decir, se alimenta un lote de datos a su red neuronal y se transforma (aumenta) aleatoriamente aplicando distintas transformaciones.

Sim embargo, desafortunadamente, **en NLP**, estas modificaciones deben realizarse con cuidado para que el texto generado siga teniendo una estructura gramátical correcta y su significado sea coherente. Además, en NLP, estos datos sintéticos deben ser generados antes del entrenamiento del modelo.

## Cosas a tener en cuenta al hacer el aumento de datos de NLP

- El conjunto de validación no debe contener datos sintéticos. 
- Si está realizando una validación cruzada de K-fold, mantener en el mismo fold, el ejemplo original y los sintéticos creados a partir de él. De esta forma, se evitará el overfitting. 
- Pruebe siempre diferentes métodos y compruebe cuál funciona mejor. Una combinación de distintas operaciones puede funcionar, pero no te excedas. A veces, estas técnicas no siempre ayudan a mejorar los resultados en NLP. 
- Prueba para determinar cuál la cantidad de ejemplos sintéticos para obtener los mejores resultados.
